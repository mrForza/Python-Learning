## Последовательности

- Поселедовательности бывают четырех типов:

    - Контейнерные - последовательности, в которых хранятся не сами объекты, а ссылки ан объекты
        - Список
        - Кортеж
        - deque

    - Плоские - последовательности, в которых объекты хранятся непосредственно друг за другом. такие виды последовательностей могут содержать в себе только объекты примитивных машинных типов 
        - str
        - bytes
        - array

    - Изменяемые - последовательности, при изменении которых под капотом не создается новый объект с обновленным состоянием
        - list
        - bytearray
        - UserList

    - Неизменяемые - последовательности, при изменении которых под капотом создается новый объект с обновленным состоянием. При этом переменная будет ссылаться на новый объект, а старый возможно будет удален сборщиком мусора
        - tuple
        - str
        - bytes
        - range
        - UserString
